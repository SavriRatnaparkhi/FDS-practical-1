```python
plt.figure(figsize=(10,7))
countries=['USA','Canada','Italy','Brazil','Thailand','India']
totalcases=[1212567,156712,2331232, 2187634,1136743,4345678]
totaldeaths=[112567,56712,231232, 148634,143743,344678]
for i in range(len(countries)):
    plt.bar(countries[i],totaldeaths[i],bottom=totalcases[i]-totaldeaths[i] ,color='black')
    plt.bar(countries[i],totalcases[i]-totaldeaths[i],color='red')
plt.xlabel('Countries')
plt.ylabel('count of cases and deaths')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
countries=['USA','Canada','Italy','Brazil','Thailand','India']
totalcases=[1212567,156712,2331232, 2187634,1136743,4345678]
totaldeaths=[112567,56712,231232, 148634,143743,344678]
totalrecovery=[437647, 89483, 239203, 284343, 53983, 789298]
plt.figure(figsize=(12,7))
x=np.arange(len(countries))
plt.bar(x, totalcases, color='red', width=0.25)
plt.bar(x + 0.25, totaldeaths, color='blue', width=0.25)
plt.bar(x + 0.5 ,totalrecovery, color='green', width=0.25)
plt.xticks([i + 0.25 for i in range(6)], countries)
plt.legend(['Total cases','Total deaths','Total recovery'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
Data={'Task':[500, 700,300],
     'Status':['Pending','In progress','Completed']}
df=pd.DataFrame(Data)
df.set_index('Status', inplace=True)
df.plot.pie(y='Task', figsize=(10,10), autopct='%1.2f%%', startangle=90)
plt.show()
```

```python
labels=['Civil','Electrical','Mechanical','Chemical']
sizes=[15,50,45,10]
fig, ax=plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.2f%%')
ax.axis('equal')
ax.set_title('Engineering Disciplines')
plt.show()
```

```python
labels=['Civil','Electrical','Mechanical','Chemical']
sizes=[15,50,45,10]
explode=[0.1,0.1,0.3,0.1]
fig, ax=plt.subplots()
ax.pie(sizes,explode=explode ,labels=labels,shadow=True, autopct='%1.2f%%')
ax.axis('equal')
ax.set_title('Engineering Disciplines')
plt.show()
```
